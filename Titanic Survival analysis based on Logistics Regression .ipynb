Drop some columns to reduce the redundancy of dataset

```python
data.drop(["Ticket","Name","Cabin"],axis=1,inplace=True)
```

Encode the data so we can use it in model.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
data["Encoded_s"]=enc.fit_transform(data["Sex"])
```

```python
data["Encoded_e"]=enc.fit_transform(data["Embarked"])
```

```python
data
```

```python
data=data.drop(["Sex","Embarked"],axis=1)
```

```python
data
```

```python
x=data.drop("Survived",axis=1)
```

```python
y=data["Survived"]
```

Apply Logistic regression to find the training and testing accuracy.

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
model=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

create a confusion matrix and create the classification report of the model.

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
import seaborn as sns
```

```python
sns.heatmap(confusion_matrix(y_test,model.predict(x_test)),annot=True)
```

```python
print(classification_report(y_test,model.predict(x_test)))
```
